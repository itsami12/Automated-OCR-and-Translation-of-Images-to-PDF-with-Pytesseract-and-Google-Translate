# convert jpg image to pdf 

```python
from PIL import Image
from reportlab.pdfgen import canvas
import glob
import os

def jpg_to_pdf(jpg_file_path, pdf_file_path):
    # Open the image using Pillow
    image = Image.open(jpg_file_path)
    
    # Get the image dimensions
    width, height = image.size
    
    # Create a canvas with the same dimensions as the image
    c = canvas.Canvas(pdf_file_path, pagesize=(width, height))
    
    # Draw the image onto the canvas at coordinates (0, 0)
    c.drawImage(jpg_file_path, 0, 0, width, height)
    
    # Save the canvas as a PDF
    c.save()
    
    print(f"Converted {jpg_file_path} to {pdf_file_path}")

# Get the list of JPEG files in the specified directory
pics = glob.glob("Telegram Desktop/*.jpg")

# Create the output directory if it doesn't exist
output_dir = "pdf_output"
os.makedirs(output_dir, exist_ok=True)

# Convert each JPEG file to a separate PDF
for file_path in pics:
    # Create the output PDF file path in the output directory
    file_name = os.path.basename(file_path)
    pdf_file_path = os.path.join(output_dir, os.path.splitext(file_name)[0] + ".pdf")
    jpg_to_pdf(file_path, pdf_file_path)
```

# convert  image to text then translate into spanish then make pdf

```python
import pytesseract
from PIL import Image
import glob
import os
from googletrans import Translator
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

translator = Translator()
language = 'es' # 'es' is the language code for Spanish

def image_to_text(image_path):
    # Open the image
    with Image.open(image_path) as img:
        # Use pytesseract to do OCR on the image
        text = pytesseract.image_to_string(img)
    return text

# Directory containing JPEG files
pics_dir = "Telegram Desktop/"

# Create the output directory if it doesn't exist
output_dir = "pdf_output"
os.makedirs(output_dir, exist_ok=True)

# Convert each JPEG file to a separate PDF
for file_path in glob.glob(os.path.join(pics_dir, "*.jpg")):
    extracted_text = image_to_text(file_path)
    translated_text = translator.translate(extracted_text, dest=language).text

    # Create PDF file name based on the original image name
    pdf_file_name = os.path.splitext(os.path.basename(file_path))[0] + "_translated.pdf"
    pdf_file_path = os.path.join(output_dir, pdf_file_name)

    # Create a PDF document
    pdf_canvas = canvas.Canvas(pdf_file_path, pagesize=letter)

    # Set font and size
    pdf_canvas.setFont("Times-Roman", 14)

    # Split text into lines
    lines = translated_text.split('\n')

    # Define the position of the text in the PDF
    y_coordinate = 750  # Starting y-coordinate
    for line in lines:
        pdf_canvas.drawString(100, y_coordinate, line)
        y_coordinate -= 14  # Move to the next line

    # Save the PDF
    pdf_canvas.save()

    print(f"PDF saved: {pdf_file_path}")
```
